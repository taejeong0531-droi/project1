```python
emotion_to_keyword = {
    "happy": "dessert",
    "sad": "comfort food",
    "angry": "spicy food",
    "tired": "coffee",
    "stressed": "chocolate",
    "neutral": "healthy"
}

def get_food_keyword(emotion):
    return emotion_to_keyword.get(emotion, "food")
```

```python
import requests

API_KEY = "87f1602c76854e1b967fe8e90e737d54"

def get_food_recommendations(keyword):
    url = "https://api.spoonacular.com/recipes/complexSearch"
    params = {
        "query": keyword,
        "number": 3,  # 추천 3개
        "apiKey": API_KEY
    }
    response = requests.get(url, params=params)
    data = response.json()

    results = []
    for item in data.get("results", []):
        results.append({
            "title": item["title"],
            "image": item["image"]
        })
    return results
```

```python
def save_to_firebase(user_id, emotion, foods):
    db.collection("recommendations").document(user_id).set({
        "emotion": emotion,
        "foods": foods
    })
```

```python
def recommend_food_by_emotion(user_id, emotion):
    keyword = get_food_keyword(emotion)
    foods = get_food_recommendations(keyword)
    save_to_firebase(user_id, emotion, foods)
    print("✅ 추천 저장 완료:", foods)

# 테스트 실행
recommend_food_by_emotion("user123", "sad")
```
